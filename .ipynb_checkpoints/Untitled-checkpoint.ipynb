```python
import pandas as pd
data = pd.read_csv('./diabetes.csv')
print(data.shape)
print(data.head)
```

```python
print(data.groupby("Outcome").size())
```

```python
X = data.iloc[:, 0:8]
Y = data.iloc[:, 8]
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
print(X_train.shape,X_test.shape)
print(Y_train.shape,Y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier, RadiusNeighborsClassifier
mod = []
for i in range(2,7):
    mod.append(("KNN_o_"+str(i),KNeighborsClassifier(n_neighbors = i)))
    mod.append(("KNN_w_"+str(i),KNeighborsClassifier(n_neighbors = i, weights = "distance")))
    mod.append(("KNN_r_"+str(i),RadiusNeighborsClassifier(n_neighbors = i, radius = 500)))

res = []
for name,model in mod:
    model.fit(X_train, Y_train)
    res.append((name,model.score(X_test, Y_test)))
for i in range(len(mod)):
    print(res[i][0],res[i][1])
from sklearn.model_selection import KFold, cross_val_score
```

```python
cv_res = []
for name, model in mod:
    kfold = KFold(n_splits = 10)
    cv_result = cross_val_score(model, X, Y, cv = kfold)
    cv_res.append((name,cv_result))
maxv = cv_res[0][1].mean()
max_num = 0
for i in range(1,len(cv_res)):
    if maxv < cv_res[i][1].mean():
        maxv = cv_res[i][1].mean()
        max_num = i
print(cv_res[max_num][0],maxv)
```

```python
knn = KNeighborsClassifier(n_neighbors = 2)
knn.fit(X_train, Y_train)
print(knn.score(X_train, Y_train),knn.score(X_test, Y_test))
```

```python
from sklearn.feature_selection import SelectKBest
sel = SelectKBest(k=2)
X_new = sel.fit_transform(X, Y)
print(X_new[0:3])
```

```python
cv_res = []
for name, model in mod:
    kfold = KFold(n_splits = 10)
    cv_result = cross_val_score(model, X_new, Y, cv = kfold)
    cv_res.append((name,cv_result))
maxv = cv_res[0][1].mean()
max_num = 0
for i in range(1,len(cv_res)):
    if maxv < cv_res[i][1].mean():
        maxv = cv_res[i][1].mean()
        max_num = i
print(cv_res[max_num][0],maxv)
```
